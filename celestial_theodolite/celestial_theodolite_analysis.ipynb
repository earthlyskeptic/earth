### Model

#### Equations for predicting declination of stars occulted by a peak for a given observer location

For an observer located at coordinates (latitude $\phi_{ob}$, longitude $\lambda_{ob}$) and a peak located at (latitude $\phi_{pk}$, longitude $\lambda_{pk}$), the central angle $\theta$ between the observer and the peak is given by:

$$\boxed{
\cos \theta = \sin(\phi_{ob}) \sin(\phi_{pk}) + 
\cos(\phi_{ob}) \cos(\phi_{pk}) \cos(\lambda_{pk} - \lambda_{ob})
}
$$

A straightforward way of deriving $\theta$ is to use an Earth-centric, Earth-fixed reference frame, take the dot product of the observer's position vector $\mathbf{r_{ob}}$ and the peak's position vector $\mathbf{r_{pk}}$ and divide the dot product $\mathbf{r_{ob} \cdot r_{pk}}$ by $(R+h) (R+H)$, where $R$ is Earth's radius, $h$ is the peak altitude and $H$ is the observer altitude (both measured above sea level).

The elevation angle of the peak as seen by the observer, $e_{pk}$, is given by: 

$$\boxed{
\sin (e_{pk}) = 
\frac{(R+h) \cos \theta - (R + H)}
{\sqrt{(R+h)^2 + (R+H)^2 - 2 (R+h) (R+H) \cos \theta}}
}
$$

This relationship can be derived by realizing that the altitude of the peak above the observer's horizontal is $(R+h) \cos \theta - (R+H)$ and the sine of $e_{pk}$ is the ratio of this altitude to the line-of-sight distance between the observer and the peak. The line of sight is given by the vector $\mathbf{v_{los}} = \mathbf{r_{pk} - r_{ob}}$, and its magnitude can be evaluated by $|\mathbf{v_{los}}| = \sqrt{\mathbf{v_{los} \cdot v_{los}}}$.

To get the azimuth of the peak from the observer, we define local east-north-up (ENU) unit vectors $\mathbf{\hat{e}}$, $\mathbf{\hat{n}}$ and $\mathbf{\hat{u}}$ at the observer, which are mutually orthogonal:

$$
\mathbf{\hat e} =  
\begin{bmatrix}
-\sin \lambda_{ob}\\
\cos \lambda_{ob}\\
0
\end{bmatrix}
\qquad
\mathbf{\hat n} = 
\begin{bmatrix}
-\sin \phi_{ob} \cos \lambda_{ob}\\
-\sin \phi_{ob} \sin \lambda_{ob}\\
\cos \phi_{ob}
\end{bmatrix}
\qquad
\mathbf{\hat u} = 
\begin{bmatrix}
\cos \phi_{ob} \cos \lambda_{ob}\\
\cos \phi_{ob} \sin \lambda_{ob}\\
\sin \phi_{ob}
\end{bmatrix}
$$

On projecting the unit line-of-sight vector $\mathbf{\hat{v}_{los} = v_{los}}/|\mathbf{v_{los}}|$ to the unit vectors along north ($\mathbf{\hat n}$) and east ($\mathbf{\hat e}$) and taking the arctangent of their ratio, we get the azimuth of the peak. $A_{pk}$:

$$
A_{pk} = \tan^{-1}
\left ( 
\frac
{\mathbf{\hat{v}_{los}} \cdot \mathbf{\hat e}}
{\mathbf{\hat{v}_{los}} \cdot \mathbf{\hat n}}
\right )
$$

On simplifying and determining the quadrant appropriately, we get:

$$
\boxed{
A_{pk} = \tan^{-1}
\left ( 
\frac
{\sin(\lambda_{pk}-\lambda_{ob}) \cos\phi_{pk}}
{\cos\phi_{ob} \sin\phi_{pk} -
\sin\phi_{ob}\,\cos\phi_{pk}\,\cos(\lambda_{pk}-\lambda_{ob})}
\right )
}
$$

The line-of-sight unit vector in ENU coordinates is:

$$
\mathbf{\hat{v}_{los}} =
\begin{bmatrix}
\cos e_{\rm pk}\,\sin A_{\rm pk} \\
\cos e_{\rm pk}\,\cos A_{\rm pk} \\
\sin e_{\rm pk}
\end{bmatrix}
$$

This can be transformed to the ECEF system to get the declination $\delta$ of the star:

$$
\boxed{
\sin\delta =
\sin\phi_{ob}\,\sin e_{\rm pk} +
\cos\phi_{ob}\,\cos e_{\rm pk}\,\cos A_{\rm pk}
}
$$

#### Equations to compute sum of squared residuals (SSR) and probability associated with a given radius value

Given the coordinates and altitudes of the observation site (latitude $\phi_{ob}$, longitude $\lambda_{ob}$, altitude $H$) and the peak (latitude $\phi_{pk}$, longitude $\lambda_{pk}$, altitude $h$) and a candidate radius $R$, we can compute the central angle $\theta$, the peak's elevation angle $e_{pk}$, the peak's azimuth $A_{pk}$ and the star's predicted declination $\delta_{predicted}$ as a function of $R$.

The expression for the weighted sum of squared residuals (SSR, $\chi^2$) is:

$$\boxed{
\chi^2(R) =
\sum_{j=1}^N
\left(
\frac{\delta_{j,\ predicted} - \delta_{j,\ cataloged}}{\sigma_j}
\right) ^2
}
$$

Here, $\delta_{j,\ predicted}$ and $\delta_{j,\ cataloged}$ are the predicted and cataologed declinations, respectively, for the $j$th observation out of $N$ observations, and $\sigma_j$ is the error associated with this observation. Since the experiment did not report errors or standard deviations, we use 0.25° or 15 arcminutes. This much error should be higher than any refraction effect at the 1-2° elevation angles observed in this experiment.

For a function of a single parameter ($R$), the 99% confidence interval is bounded by the $R$ values $R_{lower}$ and $R_{upper}$ corresponding to $\chi^2_{min} + 6.6349$.

The probability associated with a particular radius value $R$ is:

$$
L(R) = 
\left(
\sqrt{\frac{2}{\pi}}
\right)^N
\exp
\left(
{-\frac{\chi^2(R)}{2}}
\right)
$$

We can verify the 99% confidence interval by computing the probability:

$$
\dfrac{
\displaystyle \int_{R_{lower}}^{R_{upper}} L(R)
}{
\displaystyle \int_{0}^{\infty} L(R)
}
$$

```python
def deg_min_sec(angle):
    degree = np.floor(angle)
    minute = np.floor((angle - degree) * 60)
    second = ((angle - degree) * 60 - minute) * 60
    return (
        f"{degree:.0f}°{minute:.0f}\'" +
        f'{second:.1f}"'
    )


# Observer
phi_ob = 38.999700 / 180 * np.pi
lambda_ob = -104.497230 / 180 * np.pi
H = 2073.68

# Peak
phi_pk = 38.840278 / 180 * np.pi
lambda_pk = -105.044444 / 180 * np.pi
h = 4300

# Earth
R = 6371000

theta = np.arccos(
    np.sin(phi_ob) * np.sin(phi_pk) +
    np.cos(phi_ob) * np.cos(phi_pk) * np.cos(lambda_pk - lambda_ob)
)
print(f"Central angle: {deg_min_sec(theta / np.pi * 180)}")

sin_e_peak = (
    ((R + h) * np.cos(theta) - (R + H)) /
    np.sqrt((R + h)**2 + (R + H)**2 - 2 * (R + h) * (R + H) * np.cos(theta))
)
sin_e_peak = np.clip(sin_e_peak, -1.0, 1.0)
e_peak = np.arcsin(sin_e_peak)
print(
    "Peak elevation angle from observer: " +
    f"{deg_min_sec(e_peak / np.pi * 180)}"
)
```

## Results

### Result 1: For a given peak and date, occultation time of a star varies linearly as the star's right ascension, with a reciprocal slope of 15.04°/hour

A plot of occultation time (hours) vs. the (catalogued) right ascension (°) should give a straight line for each peak-date pair. The reciprocal of the slope of each line should nearly equal Earth's rotation rate of 15.04°/hour. We observe this trend for every star analyzed.

```python
# Plot right ascension vs. setting time, group by observation date
fig, ax = plt.subplots(figsize=(10, 10))
pk_dt = -1  # counter
m_inv_list = []  # list of reciprocal slopes
for (peak, date), subdf in df.groupby(level=(0, 1)):
    if len(subdf) >= 2:  # need 2 or more points for a line
        pk_dt += 1
        color = get_color(pk_dt)

        date_str = date.strftime('%b %d')
        t0 = pd.Timestamp(date)
        delta_t = df.xs((peak, date),
                        level=(0, 1))[("Observation", "Date & Time")] - t0
        hours = delta_t.dt.total_seconds() / 3600

        if hours.max() - hours.min() > 4 / 60:  # check for at least 4 min range
            ra = df.xs((peak, date), level=(0, 1))[('Star', 'RAα')]
            match (peak, date_str):  # use: angle ± 360° = same angle
                case ('Blodgett Peak', 'Feb 07'):
                    ra['LP Aquarii'] -= 360
                    ra['HD 216718'] -= 360
                case ('Pikes Peak', 'Jan 30'):
                    ra['HD 2345'] += 360

            m, b = np.polyfit(ra, hours, 1)  # best-fit line
            ra_fit = np.linspace(ra.min(), ra.max(), 100)
            hours_fit = m * ra_fit + b
            m_inv = 1 / m
            m_inv_list.append(m_inv)
            ax.plot(ra_fit, hours_fit, color=color, lw=1)

            ax.plot(ra, hours, ls='none',
                    marker='*', markersize=12, alpha=0.4,
                    mfc=color, mec='#000000', mew=0.5,
                    label=f"{peak} ({date_str}), {m_inv:.2f}°/hour")  # points

            stars = subdf.index.get_level_values(2)
            count = -1
            for xx, yy, star in zip(ra, hours, stars):
                count += 1
                if np.mod(count, 2) == 1:
                    x_disp = -3
                    y_disp = 0
                    ha = 'right'
                else:
                    x_disp = 3
                    y_disp = 0
                    ha = 'left'
                ax.text(xx + x_disp, yy + y_disp,
                        star,
                        fontsize=10,
                        color=color,
                        ha=ha)  # point labels (star names)

ax.plot([], [], '', color='none', label=' ')
ax.plot([], [],
        marker=r'$/$',
        color='black',
        ls='none',
        label=(
            f"mean reciprocal slope: {np.mean(m_inv_list):.2f}±"
            f"{np.std(m_inv_list):.2f} °/hour"
        ))
ax.legend(fontsize=12)
ax.xaxis.set_major_locator(MultipleLocator(60))
ax.yaxis.set_major_locator(MultipleLocator(1))
ax.grid(True, alpha=0.3)
plt.xlabel("Cataloged right ascension (°)", fontsize=14)
plt.ylabel("Occultation time (hours since midnight)", fontsize=14)
plt.savefig('right_ascension.jpg', dpi=1200, bbox_inches='tight')
plt.savefig('right_ascension.svg', bbox_inches='tight')
```

### Result 2: For a given peak and star, occultation time drifts linearly with observation date, with a slope of 3m:56s/day

A plot of occultation time (hours) vs. date should give a straight line for each peak-star pair. The slope of each line should nearly equal 3m:56s/day. This number is the daily stellar drift due to Earth's orbit around the Sun, and equals the ratio of the two fundamental periods of Earth---its rotation period (1 day) and its revolution period (1 year). We observe this almost exactly and uniformly for all stars.

```python
# Plot setting time vs. date, group by star
fig, ax = plt.subplots(figsize=(10, 10))
pk_st = -1  # counter
m_list = []  # list of reciprocal slopes
for (peak, star), subdf in df.groupby(level=(0, 2)):
    if len(subdf) >= 2:  # need 2 or more points for a line
        pk_st += 1
        color = get_color(pk_st)

        t0 = pd.Timestamp('Dec 31 2024')
        delta_t = df.xs((peak, star),
                        level=(0, 2))[("Observation", "Date & Time")] - t0
        days = delta_t.dt.total_seconds() / 3600 / 24
        days_frac = np.mod(days, 1)
        days_int = days - days_frac
        hours = days_frac * 24

        m, b = np.polyfit(days_int, hours, 1)  # best-fit line
        days_int_fit = np.linspace(days_int.min(), days_int.max(), 200)
        hours_fit = m * days_int_fit + b
        m_list.append(m * 60)
        ax.plot(days_int_fit, hours_fit, color=color, lw=1)

        ax.plot(days_int, hours, ls='none',
                marker='*', markersize=12, alpha=0.4,
                mfc=color, mec='#000000', mew=0.5,
                label=f"{peak}/{star}, {min_sec(m * 60)}/day")  # points

        dates = subdf.index.get_level_values(1)
        stars = subdf.index.get_level_values(2)
        for xx, yy, date, star in zip(days_int, hours, dates, stars):
            if xx < 29:
                ha = 'left'
                va = 'bottom'
            else:
                ha = 'right'
                va = 'top'
            ax.text(xx, yy,
                    f"{star} ({date.strftime('%b %d')})",
                    fontsize=10,
                    color=color,
                    ha=ha, va=va)  # point labels (star names)

ax.plot([], [], '', color='none', label=' ')
ax.plot([], [],
        marker=r'$\backslash$',
        color='black',
        ls='none',
        label=(
            f"mean slope: ({min_sec(np.mean(m_list))} ± "
            f"{min_sec(np.std(m_list))})/day"
        ))
ax.legend(fontsize=12)
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_major_locator(MultipleLocator(0.2))
ax.grid(True, alpha=0.3)
plt.xlabel("Days since Dec 31, 2024", fontsize=14)
plt.ylabel("Occultation time (hours since midnight)", fontsize=14)
plt.savefig('daily_drift.jpg', dpi=1200, bbox_inches='tight')
plt.savefig('daily_drift.svg', bbox_inches='tight')
```

### Result 3b: For a given observation site, the declination of the stars occulted by each peak can be calculated using globe geometry  

The (nearly identical) declination of the stars occulted by a peak as seen from a given observation site can be predicted by using globe geometry. This prediction uses the observer's and peak's coordinates to compute the elevation and azimuth of the peak from the observer, and then computes the declination from these angles, as described below. We see that this prediction closely matches the cataloged declination throughout the dataset.

### Equations for predicting declination of stars occulted by a peak for a given observer location

For an observer located at coordinates (latitude $\phi_{ob}$, longitude $\lambda_{ob}$) and a peak located at (latitude $\phi_{pk}$, longitude $\lambda_{pk}$), the central angle $\theta$ between the observer and the peak is given by:

$$\boxed{
\cos \theta = \sin(\phi_{ob}) \sin(\phi_{pk}) + 
\cos(\phi_{ob}) \cos(\phi_{pk}) \cos(\lambda_{pk} - \lambda_{ob})
}
$$

A straightforward way of deriving $\theta$ is to use an Earth-centric, Earth-fixed reference frame, take the dot product of the observer's position vector $\mathbf{r_{ob}}$ and the peak's position vector $\mathbf{r_{pk}}$ and divide the dot product $\mathbf{r_{ob} \cdot r_{pk}}$ by $(R+h) (R+H)$, where $R$ is Earth's radius, $h$ is the peak altitude and $H$ is the observer altitude (both measured above sea level).

The elevation angle of the peak as seen by the observer, $e_{pk}$, is given by: 

$$\boxed{
e_{pk} = 
\sin^{-1}
\left(
\frac{(R+h) \cos \theta - (R + H)}
{\sqrt{(R+h)^2 + (R+H)^2 - 2 (R+h) (R+H) \cos \theta}}
\right)
}
$$

This relationship can be derived by realizing that the altitude of the peak above the observer's horizontal is $(R+h) \cos \theta - (R+H)$ and the sine of $e_{pk}$ is the ratio of this altitude to the line-of-sight distance between the observer and the peak. The line of sight is given by the vector $\mathbf{v_{los}} = \mathbf{r_{pk} - r_{ob}}$, and its magnitude can be evaluated by $|\mathbf{v_{los}}| = \sqrt{\mathbf{v_{los} \cdot v_{los}}}$.

To get the azimuth of the peak from the observer, we define local east-north-up (ENU) unit vectors $\mathbf{\hat{e}}$, $\mathbf{\hat{n}}$ and $\mathbf{\hat{u}}$ at the observer, which are mutually orthogonal:

$$
\mathbf{\hat e} =  
\begin{bmatrix}
-\sin \lambda_{ob}\\
\cos \lambda_{ob}\\
0
\end{bmatrix}
\qquad
\mathbf{\hat n} = 
\begin{bmatrix}
-\sin \phi_{ob} \cos \lambda_{ob}\\
-\sin \phi_{ob} \sin \lambda_{ob}\\
\cos \phi_{ob}
\end{bmatrix}
\qquad
\mathbf{\hat u} = 
\begin{bmatrix}
\cos \phi_{ob} \cos \lambda_{ob}\\
\cos \phi_{ob} \sin \lambda_{ob}\\
\sin \phi_{ob}
\end{bmatrix}
$$

On projecting the unit line-of-sight vector $\mathbf{\hat{v}_{los} = v_{los}}/|\mathbf{v_{los}}|$ to the unit vectors along north ($\mathbf{\hat n}$) and east ($\mathbf{\hat e}$) and taking the arctangent of their ratio, we get the azimuth of the peak. $A_{pk}$:

$$
A_{pk} = \tan^{-1}
\left ( 
\frac
{\mathbf{\hat{v}_{los}} \cdot \mathbf{\hat e}}
{\mathbf{\hat{v}_{los}} \cdot \mathbf{\hat n}}
\right )
$$

On simplifying and determining the quadrant appropriately, we get:

$$
\boxed{
A_{pk} = \tan^{-1}
\left ( 
\frac
{\sin(\lambda_{pk}-\lambda_{ob}) \cos\phi_{pk}}
{\cos\phi_{ob} \sin\phi_{pk} -
\sin\phi_{ob}\,\cos\phi_{pk}\,\cos(\lambda_{pk}-\lambda_{ob})}
\right )
}
$$

The line-of-sight unit vector in ENU coordinates is:

$$
\mathbf{\hat{v}_{los}} =
\begin{bmatrix}
\cos e_{\rm pk}\,\sin A_{\rm pk} \\
\cos e_{\rm pk}\,\cos A_{\rm pk} \\
\sin e_{\rm pk}
\end{bmatrix}
$$

This can be transformed to the ECEF system to get the declination $\delta$ of the star:

$$
\boxed{
\sin\delta =
\sin\phi_{ob}\,\sin e_{\rm pk} +
\cos\phi_{ob}\,\cos e_{\rm pk}\,\cos A_{\rm pk}
}
$$

```python
# Plot estimated declination vs. actual declination
fig, ax = plt.subplots(figsize=(10, 10))
pk = -1  # counter

for peak in df.index.get_level_values(0).unique():
    pk += 1
    color = get_color(pk)

    dec = df.loc[peak, ("Star", "DEC")]
    mean = np.mean(dec)
    std = np.std(dec)
    num = len(dec)
    dec_range = dec.max() - dec.min()

    phi_ob = df.loc[peak, ("Observer", "Lat")] / 180 * np.pi
    lambda_ob = df.loc[peak, ("Observer", "Lon")] / 180 * np.pi
    phi_pk = df.loc[peak, ("Peak", "Lat")] / 180 * np.pi
    lambda_pk = df.loc[peak, ("Peak", "Lon")] / 180 * np.pi
    R = 6371000  # m
    h = df.loc[peak, ("Peak", "Altitude")]  # m
    H = df.loc[peak, ("Observer", "Altitude")]  # m

    theta = np.arccos(
        np.sin(phi_ob) * np.sin(phi_pk) +
        np.cos(phi_ob) * np.cos(phi_pk) * np.cos(lambda_pk - lambda_ob)
    )

    sin_e_peak = (
        ((R + h) * np.cos(theta) - (R + H)) /
        np.sqrt((R + h)**2 + (R + H)**2 - 2 * (R + h) * (R + H) * np.cos(theta))
    )
    sin_e_peak = np.clip(sin_e_peak, -1.0, 1.0)
    e_peak = np.arcsin(sin_e_peak)

    A_peak = np.atan2(
        np.sin(lambda_pk - lambda_ob) * np.cos(phi_pk),
        (
            np.cos(phi_ob) * np.sin(phi_pk) -
            np.sin(phi_ob) * np.cos(phi_pk) * np.cos(lambda_pk - lambda_ob)
        )
    )

    dec_est = np.arcsin(
        np.sin(phi_ob) * sin_e_peak +
        np.cos(phi_ob) * np.cos(e_peak) * np.cos(A_peak)
    ) / np.pi * 180

    ax.plot(dec, dec_est,
            marker='*',
            markersize=12,
            ls='none',
            mfc=color,
            mec='#000000',
            mew=0.5,
            alpha=0.4)

    match peak:  # choose label orientation
        case (
            'Cheyenne Mountain' |
            'Pikes Peak' |
            'North Peak' |
            'Mount Rosa'
        ):
            xx = mean + 2 * dec_range
            yy = mean - 2 * dec_range
            ha = 'left'
            va = 'top'
        case 'Blodgett Peak' | 'Green Mountain':
            xx = mean - 2 * dec_range
            yy = mean + 2 * dec_range
            ha = 'right'
            va = 'bottom'
        case 'Blue Mountain':
            xx = mean - 2 * dec_range
            yy = mean - 2 * dec_range
            ha = 'right'
            va = 'top'
        case _:
            xx = mean + 2 * dec_range
            yy = mean + 2 * dec_range
            ha = 'left'
            va = 'bottom'

    ax.text(xx, yy,
            (
                f"{peak}\n({num} occultations)\ndeclination: "
                f"{mean:.2f}°±{std:.2f}°"
            ),
            fontsize=10,
            ha=ha, va=va,
            color=color)

m = 1
b = 0
dec_fit = np.linspace(-50, 0, 100)
dec_est_fit = m * dec_fit + b
ax.plot(dec_fit, dec_est_fit, color='black', lw=1)

plt.xlabel("Cataloged declination (°)", fontsize=14)
plt.ylabel(("Predicted declination from observer/peak " +
            "coordinates and globe geometry (°)"), fontsize=14)
ax.set_xlim(-55, 5)
ax.set_ylim(-55, 5)
ax.xaxis.set_major_locator(MultipleLocator(10))
ax.yaxis.set_major_locator(MultipleLocator(10))
ax.grid(True, alpha=0.3)
plt.savefig('declination_correlation.jpg', dpi=1200, bbox_inches='tight')
plt.savefig('declination_correlation.svg', bbox_inches='tight')
```

### Result 3c: For a given observation site, we can use known data (declinations of the stars occulted by different peaks, observer elevation and peak elevations) to estimate the Earth's radius

For a given observation site, the relationship between the star declination and peak height uses Earth's radius as a parameter. Thus, this relationship can be used to estimate the radius by fitting a radius that minimizes the sum-of-squared residuals. We see that this estimation gives a radius of 3,911 mile, which is impressively within 1.25% of the known average radius of 3,959 mile. The 99% confidence interval for the radius is (3621 mile, 4253 mile).

### Equations to compute sum of squared residuals (SSR) and probability associated with a given radius value

Given the coordinates and altitudes of the observation site (latitude $\phi_{ob}$, longitude $\lambda_{ob}$, altitude $H$) and the peak (latitude $\phi_{pk}$, longitude $\lambda_{pk}$, altitude $h$) and a candidate radius $R$, we can compute the central angle $\theta$, the peak's elevation angle $e_{pk}$, the peak's azimuth $A_{pk}$ and the star's predicted declination $\delta_{predicted}$ as a function of $R$.

The expression for the weighted sum of squared residuals (SSR, $\chi^2$) is:

$$\boxed{
\chi^2(R) =
\sum_{j=1}^N
\left(
\frac{\delta_{j,\ predicted} - \delta_{j,\ cataloged}}{\sigma_j}
\right) ^2
}
$$

Here, $\delta_{j,\ predicted}$ and $\delta_{j,\ cataloged}$ are the predicted and cataologed declinations, respectively, for the $j$th observation out of $N$ observations, and $\sigma_j$ is the error associated with this observation. Since the experiment did not report errors or standard deviations, we use 0.25° or 15 arcminutes. This much error should be higher than any refraction effect at the 1-2° elevation angles observed in this experiment.

For a function of a single parameter ($R$), the 99% confidence interval is bounded by the $R$ values $R_{lower}$ and $R_{upper}$ corresponding to $\chi^2_{min} + 6.6349$.

The probability associated with a particular radius value $R$ is:

$$
L(R) = 
\left(
\sqrt{\frac{2}{\pi}}
\right)^N
\exp
\left(
{-\frac{\chi^2(R)}{2}}
\right)
$$

We can verify the 99% confidence interval by computing the probability:

$$
\dfrac{
\displaystyle \int_{R_{lower}}^{R_{upper}} L(R)
}{
\displaystyle \int_{0}^{\infty} L(R)
}
$$

```python
# Estimate radius from cataloged declination
def ssr(R, sigma_dec=0.25):  # sum of squared residuals
    scalar_input = np.isscalar(R)
    R = np.atleast_1d(np.asarray(R, dtype=float))
    ssr = np.zeros_like(R, dtype=float)

    for peak in df.index.get_level_values(0).unique():
        dec = df.loc[peak, ("Star", "DEC")].to_numpy(dtype=float)

        phi_ob = df.loc[peak, ("Observer", "Lat")].iloc[0] / 180 * np.pi
        lambda_ob = df.loc[peak, ("Observer", "Lon")].iloc[0] / 180 * np.pi
        phi_pk = df.loc[peak, ("Peak", "Lat")].iloc[0] / 180 * np.pi
        lambda_pk = df.loc[peak, ("Peak", "Lon")].iloc[0] / 180 * np.pi
        h = df.loc[peak, ("Peak", "Altitude")].iloc[0]  # m
        H = df.loc[peak, ("Observer", "Altitude")].iloc[0]  # m

        theta = np.arccos(
            np.sin(phi_ob) * np.sin(phi_pk) +
            np.cos(phi_ob) * np.cos(phi_pk) * np.cos(lambda_pk - lambda_ob)
        )

        sin_e_peak = (
            ((R + h) * np.cos(theta) - (R + H)) /
            np.sqrt((R + h)**2 + (R + H)**2 - 2 * (R + h) * (R + H) * np.cos(theta))
        )
        sin_e_peak = np.clip(sin_e_peak, -1.0, 1.0)
        e_peak = np.arcsin(sin_e_peak)

        A_peak = np.atan2(
            np.sin(lambda_pk - lambda_ob) * np.cos(phi_pk),
            (
                np.cos(phi_ob) * np.sin(phi_pk) -
                np.sin(phi_ob) * np.cos(phi_pk) * np.cos(lambda_pk - lambda_ob)
            )
        )

        dec_est = np.arcsin(
            np.sin(phi_ob) * sin_e_peak +
            np.cos(phi_ob) * np.cos(e_peak) * np.cos(A_peak)
        ) / np.pi * 180

        ssr += (
            ((dec[None, :] - dec_est[:, None])**2 /
             sigma_dec**2).sum(axis=1)
        )  # assuming SD of 15 arcminutes

    return ssr[0] if scalar_input else ssr


def L(R):  # probability density
    N = 0  # number of observations
    for peak in df.index.get_level_values(0).unique():
        dec = df.loc[peak, ("Star", "DEC")].to_numpy(dtype=float)
        N += len(dec)

    L = (
        np.sqrt(2 / np.pi)**N *
        np.exp(-ssr(R) / 2)
    )
    return L


# Estimate radius
res = opt.minimize_scalar(ssr)
r_est = res.x
ssr_est = res.fun

print('Estimated radius:\t\t\t',
      f'{r_est / 1000:.0f} km = {r_est / 1609.344:.0f} mile')
print('Sum of squared residuals:\t\t', f'{ssr_est:.2f}')
```

```python
# Plot SSR as a function of R
fig, ax = plt.subplots(figsize=(10, 6))

radius_fit = np.linspace(2000, 20000, 1000)  # R range in km
ssr_fit = ssr(radius_fit * 1000)

ax.plot(radius_fit, ssr_fit,
        color='r', ls='-')
ax.plot(r_est / 1000, ssr_est,
        color='k', marker='o', mfc='black')
ax.text(r_est / 1000, ssr_est + 100,
        (
            f"estimated radius: \n{r_est / 1000:,.0f} km\n" +
            f"= {r_est / 1609.344:,.0f} mile"
        ),
        ha='center', va='bottom')

plt.xlabel("R (km)", fontsize=14)
plt.ylabel("Sum of squared residuals (SSR)", fontsize=14)
ax.set_xlim(2000, 20000)
ax.grid(True, alpha=0.3)

plt.savefig('estimated_radius.jpg', dpi=1200, bbox_inches='tight')
plt.savefig('estimated_radius.svg', bbox_inches='tight')
```

```python
# Estimate 99% confidence interval (CI)
res = opt.minimize_scalar(lambda r: (ssr(r) - (ssr_est + 6.6349))**2,
                          bounds=(0.8 * r_est, r_est),
                          method='bounded')
r_lo = res.x  # lower bound of CI

res = opt.minimize_scalar(lambda r: (ssr(r) - (ssr_est + 6.6349))**2,
                          bounds=(r_est, 1.2 * r_est),
                          method='bounded')
r_hi = res.x  # upper bound of CI

print('Estimated radius:\t\t\t',
      f'{r_est / 1000:.0f} km = {r_est / 1609.344:.0f} mile')
print('99% confidence interval (CI) of radius:\t',
      f'({r_lo / 1000:.0f} km, {r_hi / 1000:.0f} km) or',
      f'({r_lo / 1609.344:.0f} mile, {r_hi / 1609.344:.0f} mile)')

# Compute probability that R falls in the CI
prob_99 = integrate.quad(L, r_lo, r_hi)[0]
prob_100 = integrate.quad(L, 0, 2e+7)[0]

print('Probability that R lies within 99% CI:\t',
      f"{prob_99 / prob_100:.4f}")
```

```python
# Plot probability density as a function of R
fig, ax = plt.subplots(figsize=(10, 6))

radius_fit = np.linspace(2000, 20000, 1000)  # R range in km
L_fit = L(radius_fit * 1000)
radius_CI = np.linspace(r_lo / 1000, r_hi / 1000, 1000)  # CI range in km
L_CI = L(radius_CI * 1000)

ax.plot(radius_fit, L_fit,
        color='blue', ls='-')
ax.plot([r_lo / 1000, r_lo / 1000], [0, L(r_lo)],
        color='k', ls='-', lw=0.5)
ax.plot([r_hi / 1000, r_hi / 1000], [0, L(r_hi)],
        color='k', ls='-', lw=0.5)
plt.fill_between(radius_CI, y1=L_CI, y2=0, where=None,
                 interpolate=True, color='k', alpha=0.1)

L_est = L(r_est)
ax.plot(r_est / 1000, L_est,
        marker='o', mec='black', mfc='red')
ax.text(r_est / 1000 + 400, L(r_est),
        (
            "estimated (most probable) R:\n" +
            f"{r_est / 1000:,.0f} km = {r_est / 1609.344:,.0f} mile"
        ),
        color='black', ha='left', va='bottom')

ax.plot(6371, L(6371 * 1000),
        marker='o', mec='black', mfc='yellow')
ax.text(6371 + 400, L(6371 * 1000),
        (
            "R known from other methods:\n" +
            f"{6371:,.0f} km = {3959:,.0f} mile"
        ),
        color='black',
        ha='left', va='top')

ax.annotate(
    '',
    xy=(r_lo / 1000, 0),
    xytext=(r_hi / 1000, 0),
    arrowprops=dict(
        arrowstyle='<->',
        connectionstyle="arc3, rad=0",
        color='blue'
    )
)

ax.annotate(
    (
        '99% confidence interval of R:\n' +
        f'({r_lo / 1000:.0f} km, {r_hi / 1000:.0f} km) or\n' +
        f'({r_lo / 1609.344:.0f} mile, {r_hi / 1609.344:.0f} mile)'
    ),
    xy=(r_est / 1000, 0),
    xytext=(1.2 * r_est / 1000, 0.2 * L_est),
    ha='left', va='top',
    arrowprops=dict(
        arrowstyle='<-',
        connectionstyle="arc3, rad=0.5",
        linestyle='--'
    )
)

plt.xlabel("R (km)", fontsize=14)
plt.ylabel("Probability", fontsize=14)

ax.set_xlim(2000, 20000)
ax.set_ylim(-0.1e-8, 3e-8)
ax.grid(True, alpha=0.3)

plt.savefig('radius_probability.jpg', dpi=1200, bbox_inches='tight')
plt.savefig('radius_probability.svg', bbox_inches='tight')
```
